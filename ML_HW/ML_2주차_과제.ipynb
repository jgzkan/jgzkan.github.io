답변: KNN은 별도의 훈련 없이 단지 데이터들 사이의 거리만을 계산한다. 즉, 훈련을 하면서 모델이 데이터에 맞게 변형되는 다른 ML 모델들과 달리 KNN은 데이터에 영향을 받지 않고 데이터를 저장해두고 거리를 계산하기만 한다.  

```python
# Code Problem 7
# 참고 데이터를 49개로 한 모델 만들기
# 참고 데이터의 개수를 바꾸는 매개변수 사용
kn49 = KNeighborsClassifier(n_neighbors = 49)
```

```python
# Code Problem 8
# kn49 모델 학습시키기
kn49.fit(fish_data, fish_target)
# kn49 모델 성능 평가하기(정확도)
kn49.score(fish_data, fish_target)
```

kn49 모델의 정확도는 (도미 개수/전체 개수)와 같습니다. 이런 현상이 나타나는 이유를 KNN이 데이터의 정답을 예측하는 과정과 정확도를 계산하는 과정을 근거로 설명하십시오.

답변: n_neighbors = 49로 설정되어있기에 KNN이 모든 fish_data의 거리를 참조한다. 그런데 도미의 개체수가 빙어의 개체수보다 많기에 kn49는 항상 도미라는 예측값을 내놓는다. 따라서 테스트하는 데이터 세트에서 도미의 경우는 모두 예측이 들어맞고 빙어는 그렇지 못한다. 그래서 kn49의 정확도가 (도미의 개체수)/(빙어의 개체수)로 나오는 것이다.

### 생각해보기 문제 3)
위 코드는 데이터의 처음 35개를 훈련 세트로 나머지 14개를 테스트 세트로 사용해서 모델을 생성하고 평가하는 과정이다. 이에 따라 정확도가 0% 가 나왔는데, 그 이유에 대해서 간단히 설명하십시오.

답변: 훈련, 테스트 데이터 세트를 기존 데이터 세트에서 섞지 않은 채 훈련/테스트 데이터 세트에 입력했기 때문에 모델은 35개의 도미 데이터 세트로 훈련을 시켰고 어떤 데이터가 들어와도 전부 도미라고 예측했다. 그런데 전부 빙어로 구성된 데이터 세트를 가지고 모델 평가를 진행하니 당연히 일치하는 예측이 없어서 성공한 예측이 없었던 것이고, 정확도가 0으로 나온 것이다. 다시 말하면, 샘플링 편향이 일어나서 정확도가 0으로 나온 것이다.

```python
import numpy as np
```

```python
# Code Problem 11
# 넘파이 함수를 이용하여 리스트처럼 한줄로 길게 출력되지 않고
# 행과 열을 맞추어 가지런히 정리된 모습으로 출력하는 코드

fish_data = np.column_stack((fish_length, fish_weight))
```

```python
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
```

```python
print(fish_target)
```

## 사이킷런으로 훈련 세트와 테스트 세트 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, random_state=42)
```

```python
print(train_input.shape, test_input.shape)
```

```python
print(train_target.shape, test_target.shape)
```

```python
print(test_target)
```

```python
# Code Problem 12
# 현재 test_target의 비율이 편향이 되어 있는 상태이다.
# 클래스 비율에 맞게 데이터를 나눠주는 매개변수를 활용해서 다시 정의하시오.

train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify=fish_target)
```

```python
print(test_target)
```

## 수상한 도미 한마리

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
print(kn.predict([[25, 150]]))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
distances, indexes = kn.kneighbors([[25, 150]])
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
